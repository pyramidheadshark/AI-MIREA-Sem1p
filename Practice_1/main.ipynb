#### 4.3.2

```python
import math
```

```python
def sample_func(x):
    return ((math.sqrt(1 + (math.e ** math.sqrt(x)) + math.cos(x ** 2))) / (abs(1 - (math.sin(x) ** 3))) + math.log(math.e, abs(2 * x)))
```

```python
func_list = [sample_func(x) for x in range(1, 21)]
func_list
```

```python
fig1, ax1 = plt.subplots()
ax1.plot(func_list, marker='o', linestyle='None')
ax1.set_title('Точечный график')
ax1.set_xlabel('Индекс')
ax1.set_ylabel('Значение')
plt.show()

fig2, ax2 = plt.subplots()
ax2.plot(func_list)
ax2.set_title('Линейный график')
ax2.set_xlabel('Индекс')
ax2.set_ylabel('Значение')
plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(func_list, marker='o', linestyle='None', label='Точечная диаграмма')
ax.plot(func_list, label='Линейная диаграмма')

ax.set_title('Результаты функции')
ax.set_xlabel('Индекс')
ax.set_ylabel('Значение')

ax.legend()

plt.show()
```

```python
def sample_func_second(x):
    return abs(math.cos(x * math.e ** (math.cos(x) + math.log(x + 1))))
```

```python
func_list_second = [sample_func_second(x) for x in range(1, 21)]
func_list_second
```

```python
area_trapz = trapezoid(func_list_second, dx=1)
```

```python
fig, ax = plt.subplots(figsize=(14, 8))

ax.plot(func_list_second, marker='o', linestyle='None', label='Точки')

x = np.arange(0, 20)
area = trapezoid(func_list_second, x, dx=1)
ax.fill_between(x, func_list_second, alpha=0.3, label=f'Площадь: {area:.2f}')

ax.set_title('Точечный график с площадью')
ax.set_xlabel('Индекс')
ax.set_ylabel('Значение')
ax.legend()

plt.xticks(range(len(func_list_second)))


plt.show()
```

```python
print("Введите 0 чтобы закончить работу калькулятора")

def e(a, b):
    return math.e ** (a + b)
def sin(a, b):
    return math.sin(a + b)
def cos(a, b):
    return math.cos(a + b)


operations = {'+': operator.add, '-': operator.sub, '*': operator.mul, '/': operator.truediv, '1': e, '2': sin,
              '3': cos, '4': operator.pow}



while True:
    s = input("""
    Введите знак(+, -, *, /) Или же номер одной из функции, где
    1. e ^ (x + y)
    2. sin(x + y)
    3. cos(x + y)
    4. x ^ y
> """)
    if s == "0":
        break
    elif s in operations:
        x = float(input("Введите x \n> "))
        y = float(input("Введите y \n> "))
        print(f"Ответ: {operations[s](x, y)}")
    else:
        print("Вы неправильно ввели операцию, попробуйте еще раз")
```
