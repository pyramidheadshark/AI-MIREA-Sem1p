#### 3.3.2

```python
url  = 'https://raw.githubusercontent.com/akmand/datasets/master/iris.csv'
data = pd.read_csv(url)
```

```python
feature_one = 'sepal_length_cm'
feature_two = 'sepal_width_cm'
features_scaled = data[[feature_one, feature_two]]
features_scaled.head()
```

```python
scaler_minmax = MinMaxScaler()
scaler_z = StandardScaler()
```

```python
features_scaled[feature_one] = scaler_minmax.fit_transform(features_scaled[[feature_one]])
features_scaled[feature_one].head()
```

```python
features_scaled[feature_two] = scaler_z.fit_transform(features_scaled[[feature_two]])
features_scaled[feature_two].head()
```
