```python
import pandas as pd
```

```python
data = pd.read_csv('data.csv')
```

```python
data.drop(columns='place', inplace=True)
```

```python
columns_mask = {
    'user_name': 'name',
    'login': 'group',
    'C(A1_Тест_1)': 'test_1',
    'D(A1_Тест_2)': 'test_2',
    'E(A1_Тест_3)': 'test_3',
    'F(A1_Тест_4)': 'test_4',
    'A(А1_Задача 1)': 'task_1',
    'B(А1_Задача 2)': 'task_2',
    'Score': 'score'
}
```

```python
data.rename(columns=columns_mask, inplace=True)
```

```python
for column in data:
    if 'task' in column or 'test' in column:
        data[column] = data[column].astype(str).str.extract(r'^(\d+)')
```

```python
data = data[data['group'].str.match(r'^[a-z]{4}-\d{2}-21-\d{2}$')]
data['group'] = data['group'].str[:-3]
data['group'] = data['group'].apply(lambda x: x.replace('bsbo', 'БСБО'))
data.sort_values(by=['group', 'name'], inplace=True)
```

```python
data.reset_index(drop=True, inplace=True)
```

```python
data.fillna(0)
```

```python
data.to_csv('data_processed.csv')
```
